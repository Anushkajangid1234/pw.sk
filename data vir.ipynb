```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

np.random.seed(42)

x = np.random.randn(1000)

y = 0.8 * x + np.random.randn(1000) * 0.2 

z = 0.3 * x + np.random.randn(1000) * 0.5  

data = pd.DataFrame({'X': x, 'Y': y, 'Z': z})

corr_matrix = data.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, vmin=-1, vmax=1)

plt.title('Correlation Matrix Heatmap')

plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import pandas as pd

np.random.seed(30)

data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}

df = pd.DataFrame(data)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['X'], df['Y'], df['Z'], c=df['Z'], cmap='viridis')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Scatter Plot of Data Points')

plt.show()
```

1. Violin Plot for Student Grades:

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

np.random.seed(15)

data_grades = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}

df_grades = pd.DataFrame(data_grades)

plt.figure(figsize=(8, 6))
sns.violinplot(x='Grade', y='Score', data=df_grades, palette='Set2')

plt.title('Distribution of Student Scores Across Different Grades')
plt.xlabel('Grade')
plt.ylabel('Score')

plt.show()
```

2. Heatmap for Sales Data Variation Across Months Days:

```python
np.random.seed(20)

data_sales = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

df_sales = pd.DataFrame(data_sales)

sales_pivot = df_sales.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

plt.figure(figsize=(10, 6))
sns.heatmap(sales_pivot, cmap='YlGnBu', annot=True, fmt='.1f', linewidths=0.5)

plt.title('Variation in Sales Across Different Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

np.random.seed(20)

data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

df = pd.DataFrame(data)

sales_pivot = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

plt.figure(figsize=(10, 6))
sns.heatmap(sales_pivot, cmap='YlGnBu', annot=True, fmt='.1f', linewidths=0.5)

plt.title('Variation in Sales Across Different Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')

plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import pandas as pd

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}

df = pd.DataFrame(data)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

np.random.seed(25)

data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))

plt.scatter(df['GDP'], df['Population'], s=df['Population']*10, alpha=0.6, c=df['Population'], cmap='viridis', edgecolors="w", linewidth=2)

plt.title('Bubble Chart: Population vs GDP')
plt.xlabel('GDP (in billion USD)')
plt.ylabel('Population (in millions)')

for i, country in enumerate(df['Country']):
    plt.text(df['GDP'][i] + 10, df['Population'][i], country, fontsize=9, ha='left')

plt.colorbar(label='Population Size')
plt.show()
```
